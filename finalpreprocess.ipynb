```python
## Geocoding (NOTE: THIS CODE SHOULDNT BE RAN UNLESS YOU WANT TO UPDATE THE CSV FILE WITH LATITUDE AND LONGITUDE DUE TO THE LIMITED QUOTA OF GOOGLE MAPS API)
## NOTE: ONLY RUN WHEN THE REST OF THE DATA IS CLEANED AND READY TO BE USED

geolocator = GoogleV3(api_key=GM_API_KEY)

    
# GeoCoding the entire address column
for index, row in df.iterrows():
    address = row['address']
    try:
        location = geolocator.geocode(address)
        print(location.address)
        print(location.latitude, location.longitude)
        
        # Use 'at' method to set values, avoiding SettingWithCopyWarning
        df.at[index, 'latitude'] = location.latitude
        df.at[index, 'longitude'] = location.longitude
    except:
        print('Error: Address not found at index ' + str(index))

df.to_csv('preprocessedData_v3.csv', index=False , encoding='utf-8-sig')
```
